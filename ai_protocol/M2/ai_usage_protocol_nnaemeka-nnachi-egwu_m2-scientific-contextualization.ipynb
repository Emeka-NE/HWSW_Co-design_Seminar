**Timestamp:** 2026-06-06T09:20:00+02:00  
**AI Tool:** Claude (claude-sonnet-4-6)


**Objective:**  
I needed to understand what clique partitioning means in the HLS context since the paper uses it for resource allocation but gives no details.

**Prompt:**
```
the paper says it uses clique partitioning for resource allocation but doesn't explain it. what does clique partitioning actually do in HLS resource allocation and how is it different from just counting operations?
```

**AI Output Summary:**  
Claude explained: in the resource allocation compatibility graph, two operations are compatible (can share a functional unit) if they are not scheduled at the same time-step and have the same type. Clique partitioning finds cliques (sets of mutually compatible ops) where each clique maps to one FU instance. For conflict graphs where ops at the same time-step conflict, the number of FUs equals the chromatic number, which for interval graphs equals the size of the maximum clique.

**Verification:**
- Level: `secondary-source-check`
- Checked: for simple time-step conflict graphs, max concurrent ops of one type = max clique = chromatic number. This is the interval graph property.
- The paper does not give details of its clique partitioning -- this interpretation is consistent with standard HLS textbooks (Teich, Gajski).
- My C++ implementation uses the max-concurrent-at-same-step approach, which is the simplest correct model.

**Evaluation:**
- Status: `accepted`
